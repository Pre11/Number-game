```python
import random
from random import randint

def fun():
    n = randint(1, 100)
    H1 = int(n/10);
    H2 = n%10;
    print("Please choose a number between 1 to 100. \n")
    i = 1
    lost = 1
    while i<=3:
        num= int(input("Enter the number: "))
        if num <= 1 or num > 100:
            print("\nPlease Enter a number between 1 to 100 ")
            continue
        elif num < n:
            print("\n The number is greater ")
            print("hint:when you perform any mathematical calacualation with the number 10,number is related to these 2 numbers "+str(H1)+", "+str(H2)+"" )
            print("you have " + str(3 - i) +  " chances left")
            i = i + 1
        elif num > n:
            print("\nthe number is smaller ")
            print("hint:when you perform any mathematical calacualation,number is related to these 2 numbers "+str(H1)+", "+str(H2)+"")
            print("you have " + str(3 - i) + " chances left")
            i = i + 1
        elif num == n:
            print("\nCongrats You have guessed the number")
            lost = 0
            break
        else:
            print("The number is invalid .\n please try again\n")
            print("you have" + str(3 - i) + " Chances left")
            continue
    if lost==1:
        print("Ooops you lost the game!!")
        print("The correct number is "+ str(n))

def main():
    print("hey user!!! \n welcome to the number game\n you have to guess a number between 1 to 100 in 3 chances else the game is lost\n good luck!!")
    fun()
    while True:
        more = input("try again.\nIf you want to play again. press 'y' = ")
        if more == "y":
            fun();
        else:
            break

main()
print("\n thankyou!!!! hope you enjoyed the game!")
```
